```python
df2= df.groupby('Country/Territory')['2022 Population'].sum().sort_values(ascending = False).head(10)
df2
```

```python
asian_countries =df[df["Continent"]=="Asia"].sort_values(by = ['2022 Population'],ignore_index=True,ascending=False)

###
europe_countries =df[df["Continent"]=="Europe"].sort_values(by = ['2022 Population'],ignore_index=True,ascending=False)

###
african_countries =df[df["Continent"]=="Africa"].sort_values(by = ['2022 Population'],ignore_index=True,ascending=False)

###
north_america =df[df["Continent"]=="North America"].sort_values(by = ['2022 Population'],ignore_index=True,ascending=False)

###
sa_countries = df.loc[df["Continent"]=="South America"].sort_values(by=["2022 Population"], ascending=False, ignore_index=True)

###
oceania =df[df["Continent"]=="Oceania"].sort_values(by = ['2022 Population'],ignore_index=True,ascending=False)
```

```python
df['Continent'].value_counts()
```

```python
sns.barplot(data = asian_countries.head(), x = 'Country/Territory' , y = '2022 Population',palette='viridis' )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title("Top 5 Countries by Population in Asian Countries")

plt.show()
```

```python
sns.barplot(data = europe_countries.head(), x = 'Country/Territory' , y = '2022 Population',palette='viridis' )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title("Top 5 Countries by Population in Europe_countries")

plt.show()
```

```python
sns.barplot(data = african_countries.head(), x = 'Country/Territory' , y = '2022 Population',palette='viridis' )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title("Top 5 Countries by Population in African_countries")

plt.show()
```

```python
sns.barplot(data = north_america.head(), x = 'Country/Territory' , y = '2022 Population',palette='viridis' )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title("Top 5 Countries by Population in north_america")

plt.show()
```

```python
sns.barplot(data = sa_countries.head(), x = 'Country/Territory' , y = '2022 Population',palette='viridis' )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title("Top 5 Countries by Population in south_america")

plt.show()
```

```python
sns.barplot(data = oceania.head(), x = 'Country/Territory' , y = '2022 Population',palette='viridis' )
plt.xlabel("Country/Territory")
plt.ylabel("2022 Population")
plt.title("Top 5 Countries by Population in oceania")

plt.show()
```

```python
continent_df = df.groupby(by='Continent').sum()
continent_df
```

```python
continent_df['2022 Population'].plot(kind='pie',figsize=(10,5), autopct='%1.1f%%')
```

```python
plt.subplots(figsize=(10,5))

trend = df.iloc[:,5:13].sum().sort_values(ascending=True)

sns.lineplot(x=trend.index, y=trend.values, marker="o")
plt.xticks(rotation=20)
plt.ylabel("Population")
plt.title("World Population Trend (1970-2022)")
plt.show()
```
